# Dataframe Bracket Selection
# We can select specific rows or columns from our DataFrame

### iloc
## The iloc indexing method selects rows and columns depending on their integer position.

## Syntax: df.iloc[row_index, column_index].

## df.iloc[0] selects the first row.

## df.iloc[:, 1] selects the second column.

## df.iloc[1:3, 0:2]: Returns a subset of rows and columns (second and third rows, first and second columns).).

# loc

## loc is used for label-based indexing, which means you may choose rows and columns based on their labels (names).

## Syntax is: df.loc[row_label, column_label].

## Example: df.loc['row_label']: This selects the row with the supplied label.

## df.loc[:, 'column_label']: This returns the column with the given label.

## df.loc['row_label1':'row_label2', 'column_label1':'column_label2']: This returns a subset of rows and columns based on labels

```python
import pandas as pd

data = {
    'A': [1, 2, 3, 4],
    'B': [5, 6, 7, 8],
    'C': [9, 10, 11, 12],
    'D': [13, 14, 15, 16]
}
names_ = pd.DataFrame(data, index=["Jack", "Thomas", "Alexandre", "Anne"])
print("Data Frame")
print(names_)
```

```python
# Let's use loc to select the first and second rows
rows_ = names_.loc[["Jack", "Thomas"]]
print(rows_)
```

```python
# let's use iloc to select the second and third columns for all rows
columns_ = names_.iloc[:, 1:3]
print(columns_)
```

```python
# let's use loc to select specific columns for all rows
columns_loc = names_.loc[:, ['B', 'C']]
print(columns_loc)
```

```python
# let's use iloc to select specific rows
# selecting the first and second rows
rows_iloc = names_.iloc[0:2] 
print(rows_iloc)
```

```python
# let use iloc to select all columns for specific rows 
# Selecting first and third rows
_rows = names_.iloc[[0, 2]]
print(_rows)
```

# 1. Mode
Replacing missing values with the mode: Mode is the most frequently occurring value in a column. This method is useful when the data is categorical.

```python
# Replace missing values with mode
# data['column_name'].fillna(data['column_name'].mode()[0], inplace=True)
```

```python
# Replace missing values with the mode for specific columns
columns_with_missing_values = ['Sweetener', 'Biscuits', 'Yoghurt']
```

```python
for column in columns_with_missing_values:
    mode_value = food[column].mode()[0]
    food[column].fillna(mode_value, inplace=True)
```

# Removing outliers and saving the data in our dataset using z-score

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
food_ = pd.read_csv("C:/Users/user/Downloads/food-consumption.csv")
print(food_.head())
```

```python
# calculating Z-scores for each column except "Country"
# np.abs() takes the absolute value of the z-scores.
cols = food_.columns[1:]
z_scores = np.abs(stats.zscore(food_[cols]))
```

```python
# Identifying and Removing Outliers
# z-score > 2 for extreme outliers and filtering the dataset
# .any(axis=1) checks each row if it has at least one outlier.
threshold = 2
outliers = (z_scores > threshold).any(axis=1)
cleaned_data = food_[~outliers]
```

```python
print(cleaned_data)
```

```python
# Dropping rows with any NaN values
cleaned_data = cleaned_data.dropna(axis=0, how='any')
```

```python
cleaned_data
```

```python
# Boxplot before outlier removal
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.boxplot(food_.iloc[:, 1:].values, labels=food_.columns[1:], vert=False)
plt.title('Before Outlier Removal')
plt.xlabel('Values')
plt.ylabel('Features')


# Boxplot after outlier removal
plt.subplot(1, 2, 2)
plt.boxplot(cleaned_data.iloc[:, 1:].values, labels=cleaned_data.columns[1:], vert=False)
plt.title('After Outlier Removal')
plt.xlabel('Values')
plt.ylabel('Features')

plt.tight_layout()
plt.show()
```

```python
# Let's save the cleaned data
cleaned_data.to_csv("cleaned_data.csv", index=False)
```
